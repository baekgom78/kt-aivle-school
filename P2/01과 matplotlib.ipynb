## 1.라이브러리 불러오기

```python
import matplotlib.pyplot as plt
```

* 데이터 시각화

> 수많은 데이터를 한눈에 파악할 수 있는 방법  
① 시각화 도구 : histogram, boxplot, densityplot, scatter, ...  
③ 수치화 도구 : min, max, mean, std, 사분위수, ...


* Focus
    * 우리가 다루는 데이터는 비즈니스를 담고 있습니다.
    * 데이터 시각화의 목적은 
    * 아름다운 그래프가 아니라, 
    * 통계적인 해석을 넘어,
    * 한눈에 파악해야 할 정보는 **비즈니스의 인사이트** 입니다.

* 한계
    * 그래프와 통계량은 대부분 정보(원본)를 요약합니다.

    * 어떤 식으로든 요약을 하면, 정보의 손실이 발생 됩니다.

## 4.추가기능①

## 5.추가기능② : 여러 그래프 나눠서 그리기

![](https://codetorial.net/matplotlib/_images/subplot_03.png)

* row : 고정된 행 수
* column : 고정된 열 수
* index : 순서 (1부터 시작)

#### 1) 위, 아래로(2행 1열) 그래프 그리기.

```python
dict1 = {'v1': [1,2,3,4,5], 'v2': [2,5,3,1,2], 'v3':[4,6,7,6,5]}

plt.subplot(2,1,1)
plt.plot('v1', 'v2', 'go--', data = dict1)

plt.subplot(2,1,2)
plt.plot('v1', 'v3', 'rs-', data = dict1)

plt.tight_layout() # 그래프간 간격을 적절히 맞추기
plt.show()
```

#### 2) 옆으로 (1행 2열) 그래프 그리기.

```python
dict1 = {'v1': [1,2,3,4,5], 'v2': [2,5,3,1,2], 'v3':[4,6,7,6,5]}

plt.subplot(1,2,1)
plt.plot('v1', 'v2', 'go--', data = dict1)

plt.subplot(1,2,2)
plt.plot('v1', 'v3', 'rs-', data = dict1)

plt.tight_layout() # 그래프간 간격을 적절히 맞추기
plt.show()
```

### 2) 추가 기능

① 다음의 조건에 따라 라인차트를 그리기 위한 코드를 작성하고, 축의 범위를 조정해 봅시다.
* 데이터
    * 'v1': [1, 2, 3, 4, 5]
    * 'v2': [2, 5, 3, 1, 2]
    * 'v3': [4, 6, 7, 6, 5]
* v1, v3의 값으로 지정
* 레이블: mango
* color: red
* marker: square
* line style: solid
* 축의 범위
    * x축: (0, 6)
    * y축: (0, 8)
* 그리드 / 레이블 추가

```python
dict1 = {'v1': [1, 2, 3, 4, 5], 'v2': [2, 5, 3, 1, 2], 'v3': [4, 6, 7, 6, 5]}

plt.plot('v1','v3','rs-',data=dict1,label='mango')
plt.xlim(0,6)
plt.ylim(0,8)
```

② 위에서 만들었던 그래프의 크기를 조정해 봅시다.
* 그래프 크기
    * 가로: 8
    * 세로: 6

```python
dict1 = {'v1': [1, 2, 3, 4, 5], 'v2': [2, 5, 3, 1, 2], 'v3': [4, 6, 7, 6, 5]}

plt.plot('v1','v3','rs-',data=dict1,label='mango')
plt.xlim(0,6)
plt.ylim(0,8)
plt.figure(figsize = (8, 6))
```

③ 다음의 조건에 따라 라인차트를 그리기 위한 코드를 작성하고, 수평선, 수직선을 추가해 봅시다.  

단, 라인스타일, 마커, 칼라 등은 자유롭게 조정해도 됩니다.
* 데이터
    * 'v1': [1, 2, 3, 4, 5]
    * 'v2': [2, 5, 3, 1, 2]
    * 'v3': [4, 6, 7, 6, 5]
* v1, v3의 값으로 지정
* 레이블: mango
* color: red
* marker: square
* line style: solid
* 수평선
    * y값: 5.3
    * color: grey
    * line style: dashed
* 수직선
    * x값: 2.4
    * color: grey
    * line style: dashed

```python
dict1 = {'v1': [1, 2, 3, 4, 5],
'v2': [2, 5, 3, 1, 2],
'v3': [4, 6, 7, 6, 5]}

plt.plot('v1','v3','rs-',data=dict1,label='mango')
plt.axvline(2.4,color='grey',linestyle='dashed')
plt.axhline(5.3,color='grey',linestyle='dashed')
```

④ 위에서 만들었던 그래프에 텍스트를 추가해 봅시다.
* (2.5, 4.2)에 수직선의 값을 나타내는 텍스트 추가
* (1.2, 5.4)에 수평선의 값을 나타내는 텍스트 추가

```python
dict1 = {'v1': [1, 2, 3, 4, 5],
'v2': [2, 5, 3, 1, 2],
'v3': [4, 6, 7, 6, 5]}

plt.plot('v1','v3','rs-',data=dict1,label='mango')
plt.axvline(2.4,color='grey',linestyle='dashed')
plt.axhline(5.3,color='grey',linestyle='dashed')
plt.text(2.5, 4.2, 2.4)
plt.text(1.2, 5.4, 5.4)
```

⑤ 다음의 그래프들을 위아래로 나누어 그리시오.

단, 라인스타일, 마커, 칼라 등은 자유롭게 조정해도 됩니다.
* 데이터
    * 'v1': [1, 2, 3, 4, 5]
    * 'v2': [2, 5, 3, 1, 2]
    * 'v3': [4, 6, 7, 6, 5]
* 그래프 1
    * v1, v2의 값으로 지정
    * color: green
    * marker:  circle
    * line style: dashed
* 그래프 2
    * v1, v3의 값으로 지정

```python
dict1 = {'v1': [1, 2, 3, 4, 5],
'v2': [2, 5, 3, 1, 2],
'v3': [4, 6, 7, 6, 5]}
plt.subplot(2,1,1)
plt.plot('v1','v2','go:', data=dict1)
plt.subplot(2,1,2)
plt.plot('v1','v3','rs-', data=dict1)
```

⑥ 다음의 그래프들을 좌우로 나누어 그리시오.

단, 라인스타일, 마커, 칼라 등은 자유롭게 조정해도 됩니다.
* 데이터
    * 'v1': [10, 20, 30, 40, 50]
    * 'v2': [15, 40, 25, 30, 10]
    * 'v3': [30, 20, 15, 25, 15]
* 그래프 1
    * v1, v2의 값으로 지정
    * color: yellow
    * marker:  square
    * line style: dotted
* 그래프 2

```python
dict1 = {'v1': [10, 20, 30, 40, 50],
'v2': [15, 40, 25, 30, 10],
'v3': [30, 20, 15, 25, 15]}
plt.subplot(1,2,1)
plt.plot('v1','v2','ys:', data=dict1)
plt.subplot(1,2,2)
plt.plot('v1','v3','co--', data=dict1)
```
